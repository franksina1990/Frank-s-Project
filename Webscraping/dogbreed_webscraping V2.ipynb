```python
import webbrowser 
import pyautogui as pg
import time
import pyperclip 
import re
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

# Select features need to scrap. 
featuresname= ['Energy Level','Exercise Requirements', 'Playfulness', 'Affection Level', 'Friendliness to Dogs', 'Friendliness to Other Pets', 
              'Friendliness to Strangers','Watchfulness','Ease of Training', 'Grooming Requirements', 'Heat Sensitivity','Vocality']

datadict={'Breed':featuresname}
    
originaldf=pd.DataFrame.from_dict(datadict)
    
# Read the website and paste the html by hotkey. soup.get_text didn't work for this website. 
def readhtml(url):
    webbrowser.open(url)
    time.sleep(5)
    pg.hotkey('ctrl','u')
    time.sleep(2)
    pg.hotkey('ctrl','a')
    pg.hotkey('ctrl','c')
    time.sleep(2)
    pg.hotkey('ctrl','w')
    pg.hotkey('ctrl','w')
    
    dd=pyperclip.paste()
    return dd

# Scrap data from website by using beautifulsoup. 
def mysoup(dd):
    soup = BeautifulSoup(dd,'html.parser')
    
    a1=soup.find_all("span",class_="u-isVisuallyHidden")
    
    p= [a2.get_text() for a2 in a1]
    
    mylist=[]
    for num in range (6,18):
        mylist.append(p[num])
    
    
    featureslist=[]
    for n in range (0,12):
        fn=mylist[n][30:31]
        featureslist.append(fn)
        
    return featureslist

# Create dataframe. 
def createdf(featureslist):     
    featureslist=['Spinone-Italiano']+featureslist
    
    featuresname= ['breed','Energy Level','Exercise Requirements', 'Playfulness', 'Affection Level', 'Friendliness to Dogs', 'Friendliness to Other Pets', 
              'Friendliness to Strangers','Watchfulness','Ease of Training', 'Grooming Requirements', 'Heat Sensitivity','Vocality']
    
    datadict={'Features Names':featuresname,'features':featureslist}
    
    dataframe=pd.DataFrame.from_dict(datadict)
    return dataframe

# Add columns
def addcolumns(featureslist):       
        
    originaldf['Labrador-Retriever']=featureslist
    return originaldf   
```

```python
# Read HTML and store to dateframe. If cannot find the breed, print alarm. 
dogtable=pd.read_csv("uniquedogs.csv")
urllist=list(dogtable['Url'])
breedname= list(dogtable['breed'])
namelen=len(breedname)

for n in range (0,namelen):
    try: 
        dd=readhtml(urllist[n])
        time.sleep(2)
        featureslist=mysoup(dd)
        originaldf[breedname[n]]=featureslist
    except: 
        print("Breed {} doesn't have record on petfinder.com".format(breedname[n]))

originaldf
```
