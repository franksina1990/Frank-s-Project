```python
import pandas as pd
import numpy as np
```

```python
# Load dog_mix_breed csv.
dog_mix_breed = pd.read_csv("dog_mix_breed.csv")
dog_mix_breed.head()
```

```python
#Load dog_features2 csv. 
dog_features = pd.read_csv("doggy_features2.csv")
dog_features.head()
```

```python
# Remove "_" in each dog breed. 
features_columns=dog_features.columns
dog_features.columns=features_columns.str.replace('-',' ')
```

```python
# Swapping dag_features
dog_features_swapped= dog_features.T
```

```python
# Deal with Header
dog_features_swapped.columns=dog_features_swapped.iloc[0,:]
dog_features_swapped=dog_features_swapped.drop("Breed")
dog_features_swapped.head()
```

```python
# Make a copy of dog_features_list columns 
dog_features_list=dog_features_swapped.index.to_list()
dog_features_swapped_columns=dog_features_swapped.columns.to_list()
```

```python
# Add dog_features_list columns to dog_mix_breed
for column in dog_features_swapped_columns:
    dog_mix_breed[str(column)] = ""
dog_mix_breed.head()
```

```python
# For loop to calculate the mix breed features (Average of breeds in mix breeds)
for i in range(0,len(dog_mix_breed)):
    mylist=[]
    for breed in dog_features_list:
        if dog_mix_breed.iloc[i].str.contains(str(breed))[0]==True:
            mylist.append(breed)

    n=0
    for column in dog_features_swapped_columns:        
        dog_mix_breed[column][i]=dog_features_swapped[dog_features_swapped.index.isin(mylist)].iloc[:,n].mean()
        n+=1
dog_mix_breed.tail()
```

```python
dog_features_swapped2=dog_features_swapped
#dog_features_swapped2.to_csv("dog_features_swapped2.csv")
dog_features_swapped2.insert(loc=0, column='breed', value=dog_features_swapped2.index)
#dog_features_swapped2.reset_index()
dog_features_swapped2.head()
```

```python
dog_mix_breed_backup=dog_mix_breed
#dog_mix_breed_backup.reset_index(drop=True, inplace=True)
#dog_mix_breed_backup.to_csv("dog_mix_breed3.csv")
dog_mix_breed_backup.tail()
```

```python
# Concat two csv.
all_dog_breed = pd.concat([dog_mix_breed_backup, dog_features_swapped2])
```

```python
#Reset index
all_dog_breed.reset_index().drop(columns='index')
```

```python
all_dog_breed.dropna(inplace=True)
```
